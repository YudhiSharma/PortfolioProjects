**Gross is a numerical value but the data type is presented as object. Hence, converting data type for better data visualization.

```python
df['Gross'] = df['Gross'].str.replace(',', '')
df['Gross'] = df['Gross'].astype('float64')
df['Gross'] = df['Gross'].replace(np.nan, 0)
```

```python
df['Gross'] = df['Gross'].astype(int)
```

## Appearance of Stars in movies (Top Voted, Top Grossing, Top Rated, Highest Meta Score)

```python
stars=['Star1','Star2','Star3','Star4']
fig,axs=plt.subplots(4,1,figsize=(20,10))
ax=0
for x in stars:
    s=df.groupby([x]).sum().reset_index()
    d=s.sort_values(['No_of_Votes'],ascending=False)[:10]
    axs[ax].bar(d[x],d['No_of_Votes'])
    axs[ax].set_title(x)
    axs[ax].set_ylabel("Appearances", weight = "bold")
    ax+=1
    plt.tight_layout()
```

### Certificate Count

```python
df['Certificate']=df['Certificate'].fillna('Not Rated')

certificates=df[df['Certificate']!='Not Rated']['Certificate']
fig,axs=plt.subplots(figsize=(20,10))
g=sns.countplot(certificates)
g.set_title("Count of Certificates provided", weight = "bold")
plt.show()
```

## TOP Genres

```python
from collections import Counter
genre=[]
for x in df['Genre']:
    for y in x.split(','):
        genre.append(y.strip().lower())

count=Counter(genre)
count=count.most_common()[:10]
x,y=map(list,zip(*count))
fig,axs=plt.subplots(figsize=(20,5))
g=sns.barplot(y,x)
g.set_ylabel("Genres", weight = "bold")
g.set_title("Top Ten Genres", weight = "bold")
plt.show()
```
